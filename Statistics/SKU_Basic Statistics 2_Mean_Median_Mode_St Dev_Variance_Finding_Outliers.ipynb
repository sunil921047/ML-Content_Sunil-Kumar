## Basic Concept of Statistics

When performing statistical analysis on a set of data, the mean, median, mode, and standard deviation, etc are all helpful values to calculate. 

###### Mean

The Mean is also known as the 'average'. We can get the Mean by summing all the data values and then divided by it with total numbers of data points.

The mean is the sum of all the values in the data divided by the total number of values in 
the data.

You can calculate the mean for both the population and the sample.

Calculating the population mean (Î¼) and sample mean (x) is done with this formula:


```python
values = [2,4,5,2,8,3,12]
import numpy as np
mean = np.mean(values)
mean
```

###### Median

The Median is the middle value of the given dataset if total data points is odd number. Or the average of two middle values if thetotal data points is even number.
It is highly important when seperating the dataset into two equal bins.


```python
values = [2,4,5,2,8,3,12]
import numpy as np
mean = np.median(values)
mean
```

###### Mode

The mode of the dataset is the datapoint which occurs maximum times.

Mode is the most repetitive value of a given set of values.

```python
values = [2,4,5,2,8,3,12]
import statistics as stat
mean = stat.mode(values)
mean
```

##### Variance

Variance is a measure of how data points differ from the mean.

Variance is use to get how far the data points are from the mean value.

Variance is the expected value of the squared variation of a random variable from its mean value, in probability and statistics. Informally, variance estimates how far a set of numbers (random) are spread out from their mean value

###### Covariance

###### Standard Deviation

Standard deviation is use to get the dispersion of datapoints from the mean in a given dataset.

It is the square root of variance of dataset.

```python
import numpy as np
items = (2,4,5,6,4,7,8,9,1,32)
st_dev = np.std(items)
st_dev
```

### Measures of dispersion

###### Range

Range is the number which we get after subtracting the maximum value with minimum value of given dataset.


```python
items = (2,4,5,6,4,7,8,9,1,32)
my_min = min(items)
my_max = max(items)
my_range = my_max-my_min
print('Range of given dataset is :  ' , my_range)
```

###### Quartiles

The quartiles data into three pointsâa lower quartile, median, and upper quartileâto form four groups of the dataset.

Quartiles are used to calculate the interquartile range, which is a measure of variability around the median.

###### Lower quartile (first quartile)

The lower quartile is the value under which 25% of data points are found when they are arranged in increasing order. 

First Quartile(Q1) = ((n + 1)/4)th Term 

```python
items = [49,47,45,6,43,7,15, 36,39,41,41,43]
items.sort()
items
```

```python
lower_quar = len(items)/4
print('Lower Quartile Number : ' , lower_quar)
print ('Lower Quartile Data Point : ' , items[3-1])
Q1=items[3-1]
Q1
```

###### Upper quartile (third quartile)

The upper quartile, or third quartile (Q3), is the value under which 75% of data points are found when arranged in increasing order.

```python
upper_quar = (len(items)/4)*3
print('Upper Quartile Number : ' , upper_quar)
print ('Upper Quartile Data Point : ' , items[9-1])
Q3 = items[9-1]
Q3
```

###### Middle Quartile(second quartile or Median)

```python
middle_quar = (len(items)/2)
print('Middle Quartile Number  : ' , middle_quar)
print ('Middle Quartile or Median Data Point : ' , items[6-1])
Q2 = items[6-1]
```

###### inter Quartile Range (IQR)

The interquartile range is calculated as Upper Quartile â Lower Quartile.

```python
IQR = Q3 - Q1 
print('Inter Quartile Range of given dataset : ', IQR)
```

###### Quartile Deviation

Quartile deviation is half of the distance between the third and the first quartile. It is also known as the Semi Interquartile range.

```python
Q_Deviation = (Q3 - Q1)/2
print('Quartile Deviation of given dataset : ' , Q_Deviation)
```

## What is Outlier

An outlier is an observation which lies an abnormal distance from other datapoints in a given dataset.

An outlier is an observation that lies an abnormal distance from other 
values in a random sample from a population

An outlier is a data point that lies outside the overall pattern in a distribution.

The Lower fence is the "lower limit" and the Upper fence is the "upper limit" of data, and any data lying outside this defined bounds can be considered an outlier.

The Lower fence is the 'lower limit' and the Upper fence is the 'upper limit' of data, and any data which does not fit in the upper and lower limits (fence) is considered as the outliers.
