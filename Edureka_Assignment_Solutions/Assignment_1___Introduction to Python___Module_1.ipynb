### Question 3
Write a program that will find all the numbers between 1000 and 3000 (both
excluded) such that each digit of a number is an odd number. Print the number
of such elements

```python
st_num = 2000
end_num = 3000
list1=[]
list2=[]

for i in range (st_num+1  , end_num):
    a = str(i)
    
    if (int(a[0]) %2 == 0) and (int(a[1]) %2 == 0) and (int(a[2]) %2 == 0) and (int(a[3]) %2 == 0):
        list1.append(a)
    else:
        list2.append(a)
       
print(('All digits of this number is odd : ' , list1 ))
print(('All digits of this number is even : ' , list2 ))
```

##  Question 4
Write a program that accepts a stringand calculatesthe number of letters and digits.Suppose if the entered string is: Edureka123 

```python
my_str = 'Edureka123'
count = 0
import re
num = re.finditer('[0-9]',my_str)

for i in num:
    count=count+1

print('Tota Alphabets used in given string : ' , len(my_str) - count)
print(('Total Numeric used in given string : ' ), count)
```

```python
my_str = 'Edureka123@#!'
count = 0
import re
letter = re.finditer('[\D]', my_str)

for i in letter:
    count = count + 1
print  ('Totol LETTERS : ' , len(my_str) - count)
print ('Total DIGITS : ', count)
```

```python
my_str = 'Edureka123@#'
count = 0
import re
letter = re.finditer('[\d]' , my_str)

for i in letter:
    count = count + 1
print  ('Totol LETTERS : ' , len(my_str) - count)
print ('Total DIGITS : ', count)
```
