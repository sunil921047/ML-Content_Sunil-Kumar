# Dummy Variables 

```python
import pandas as pd
import numpy as np   
```

```python
City = ['Meerut' , 'Meerut' , 'Meerut' , 'Meerut' , 'Meerut' , 'Baghpat' , 'Baghpat' , 'Baghpat' , 'Baghpat' , 'Khekra' , 'Khekra' , 'Khekra' , 'Khekra']
Price = [550000,565000,610000,680000,725000,585000,615000,650000,710000,575000,600000,620000,695000]
Area = [2600,3000,3200,3600,4000,2600,2800,3300,3600,2600,2900,3100,3600]
 
df= pd.DataFrame({'City' : City , 
                 
                 'Area' : Area,
                 'Price' : Price})
```

```python
df.head(3)
```

###### We have one column with Categorical Data - City

###### Transform the Categorical Data in Numerical Data with - Dummy Variable

```python
pd.get_dummies(df)
```

```python
import pandas as pd
import numpy as np
data = {"col1":["A","B","C","D"],"col2":["E","D","F","G"],"col3":[1,2,3,4]}
df = pd.DataFrame(data=data)
```
