# Creating a Linear Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
# https://scikit-learn.org/stable/index.html
```

```python
df = pd.read_excel('Linear_Reg_Data.xlsx')
df.head(2)
```

```python
df1 = df[['Expenses' , 'Profit']]
df1.head(2)
```

```python
# data in series
x = df.loc[: , 'Expenses']
y=df.loc[ : , 'Profit']
```

```python
# data transfer in DataFrame
x = pd.DataFrame(x)
y = pd.DataFrame(y)
```

```python
#import libraries
from sklearn.model_selection import train_test_split
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(x , y , train_size = 0.8, random_state=45 )
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_Train, Y_Train)
```

```python
prediction = lin_reg.predict(Y_Test)
```

```python
Y_Test
```

```python
Y_Test['Predict_Profit'] = prediction
Y_Test
```

```python
from sklearn.metrics import r2_score
```

```python
#accuracy = r2_score(X_Test ,X_Test['Prediction'] )
```

```python
accuracy = r2_score(Y_Test['Profit'] , Y_Test['Predict_Profit'])
accuracy
```
