# How to calculate Slope and Intercept

### Slope

The slope of a line calculates the "steepness" of a line. It is usually denoted by the letter m. So, the slope of a line is the change in Y divided by the change in X

The slope of a line is also defined as the ratio of rise over run. 

y = m + bx

m is the slope of the line

b is the y-intercept of the line
