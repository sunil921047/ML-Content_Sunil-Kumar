## How to reverse the rows and save with the index

```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry]')
drinks.head(3)
```

```python
# Basically we are giving -1 in place of Step
drinks.loc[::-1].head(6)
```

```python
# Basically we are giving -1 in place of Step
drinks.loc[::-2].head(5)
```

###### If we want to store data as same but with new index

```python
drinks.loc[::-1].reset_index(drop=True).head(5)
```

## How to reverse the columns and save with the index

```python
drinks.loc[::,::-1].head(5)
```
